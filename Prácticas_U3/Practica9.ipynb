```python
stopwords = [
    'a', 'al', 'antes', 'como', 'con', 'de', 'del', 'desde', 'durante',
    'el', 'en', 'era', 'eran', 'es', 'estas', 'fue', 'fueron', 'ha',
    'han', 'hasta', 'la', 'las', 'le', 'lo', 'los', 'o', 'otra', 'otro',
    'para', 'por', 'que', 'quienes', 'se', 'si', 'sin', 'su', 'sus',
    'también', 'un', 'una', 'y', 'no', 'más', 'pero', 'ser', 'son',
    'esto', 'este', 'esta', 'hay', 'ni', 'cual', 'cuyo', 'cuya'
]
```

# Preprocesamiento

```python
def es_valido(caracter):
    c = ord(caracter)
    if (c >= 65 and c <= 90) or (c >= 97 and c <= 122):
        return True
    vocales_tildes_n = [
        225, 233, 237, 243, 250, 241,   # minuscula
        193, 201, 205, 211, 218, 209,   # mayusculas
        252, 220                         # dieresis
    ]
    return c in vocales_tildes_n

def to_minusculas(texto):
    letras = ""
    for letra in texto:
        if ord(letra) >= 65 and ord(letra) <= 90:
            letras += chr(ord(letra) + 32)
        else:
            letras += letra
    return letras

def tokenizador(texto, stopwords):
    token = ""
    tokens = [None] * len(texto)
    j = 0
    for i in range(len(texto)):
        if es_valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                if token not in stopwords:
                    tokens[j] = token
                    j += 1
                token = ''
    return tokens[0:j]
```

```python
# Separar el texto en oraciones para el resumen final
def dividir_oraciones(texto):
    # Separa por punto, exclamacion o interrogacion
    oraciones = []
    actual = ""
    for c in texto:
        actual += c
        if c in ['.', '!', '?'] and len(actual.strip()) > 15:
            oraciones.append(actual.strip())
            actual = ""
    return [o for o in oraciones if len(o) > 20]

oraciones_originales = dividir_oraciones(texto)
print(f"Total de oraciones detectadas: {len(oraciones_originales)}")
print()
print("Primeras 3 oraciones:")
for i in range(3):
    print(f"  [{i}] {oraciones_originales[i][:85]}...")
```

```python
# Tokenizar el texto completo (para entrenar Word2Vec)
tokens_global = tokenizador(to_minusculas(texto), stopwords)

print(f"Tokens del texto completo: {len(tokens_global)}")
print(f"Primeros 15 tokens: {tokens_global[:15]}")
```

# Vocabulario
* Construir word2idx (palabra → índice) y idx2word (índice → palabra)

```python
# Palabras únicas
vocab_lista = []
for t in tokens_global:
    ya_esta = False
    for v in vocab_lista:
        if v == t:
            ya_esta = True
            break
    if not ya_esta:
        vocab_lista.append(t)

vocab_size = len(vocab_lista)

# Diccionarios de traducción
word2idx = {}
idx2word = {}
for i in range(vocab_size):
    word2idx[vocab_lista[i]] = i
    idx2word[i] = vocab_lista[i]

print(f"Tamaño del vocabulario: {vocab_size} palabras")
print(f"Primeras 10: {vocab_lista[:10]}")
```

Se toman las palabras dentro de la ventana de contexto como objetivo. Asi se le enseña al modelo qué palabras suelen aparecer juntas.

```python
window_size = 2

X_idx = []   # índices de palabras centrales
y_idx = []   # índices de palabras de contexto

for i in range(len(tokens_global)):
    central_idx = word2idx[tokens_global[i]]
    inicio = max(0, i - window_size)
    fin    = min(len(tokens_global), i + window_size + 1)

    for j in range(inicio, fin):
        if i != j:
            X_idx.append(central_idx)
            y_idx.append(word2idx[tokens_global[j]])

print(f"Pares Skip-gram generados: {len(X_idx)}")
print(f"Ejemplo: palabra '{idx2word[X_idx[0]]}' → contexto '{idx2word[y_idx[0]]}'")
```

Convertir cada indice a un vector one-hot

```python
def one_hot(indice, tamaño):
    vector = np.zeros(tamaño)
    vector[indice] = 1
    return vector

X_oh = np.array([one_hot(i, vocab_size) for i in X_idx], dtype=np.float32)
y_oh = np.array([one_hot(i, vocab_size) for i in y_idx], dtype=np.float32)

print(f"Shape X_oh: {X_oh.shape}")
print(f"Shape y_oh: {y_oh.shape}")
```

# Modelo Word2Vec

```python
embedding_dim = 30   # dimensión del vector de cada palabra

W1 = tf.Variable(tf.random.normal([vocab_size, embedding_dim]))
W2 = tf.Variable(tf.random.normal([embedding_dim, vocab_size]))

def forward(x):
    hidden = tf.matmul(x, W1)   # one-hot × W1 = embedding de la palabra
    logits = tf.matmul(hidden, W2)  # embedding × W2 = puntuaciones por palabra
    return logits

def loss_fn(logits, labels):
    # Entropía cruzada con softmax: compara distribución predicha vs. one-hot real
    return tf.reduce_mean(
        tf.nn.softmax_cross_entropy_with_logits(labels, logits)
    )
```

# Entrenamiento

```python
optimizer   = tf.optimizers.Adam(0.01)
epochs      = 500
loss_history = []

for epoch in range(epochs):
    with tf.GradientTape() as tape:
        logits = forward(X_oh)
        loss   = loss_fn(logits, y_oh)

    grads = tape.gradient(loss, [W1, W2])
    optimizer.apply_gradients(zip(grads, [W1, W2]))
    loss_history.append(loss.numpy())

    if epoch % 100 == 0:
        print(f"Epoch {epoch:>4}   Loss: {loss.numpy():.4f}")
```

```python
# Los embeddings quedan en W1
embeddings = W1.numpy()
print(f"Matriz de embeddings: {embeddings.shape}  ({vocab_size} palabras × {embedding_dim} dims)")
```

# Similitud coseno y palabras similares


```python


def palabras_similares(palabra, top_n=5):
    idx     = word2idx[palabra]
    vector  = embeddings[idx]
    similitudes = []
    for i in range(vocab_size):
        if i != idx:
            sim = similitud_coseno(vector, embeddings[i])
            similitudes.append((idx2word[i], sim))
    # Ordenar descendente
    for i in range(len(similitudes) - 1):
        for j in range(i + 1, len(similitudes)):
            if similitudes[j][1] > similitudes[i][1]:
                similitudes[i], similitudes[j] = similitudes[j], similitudes[i]
    return similitudes[:top_n]
```

# Resumen automático con Word2Vec

El vector de una oración = promedio de los embeddings de sus palabras.

La oración más cercana (por similitud coseno) al vector del documento completo es la más representativa.

```python
def similitud_coseno(a, b):
    numerador   = np.dot(a, b)
    denominador = np.linalg.norm(a) * np.linalg.norm(b)
    if denominador == 0:
        return 0.0
    return numerador / denominador

def vector_oracion(tokens):
    # Promedio de embeddings de los tokens que están en el vocabulario
    vecs = []
    for t in tokens:
        if t in word2idx:
            vecs.append(embeddings[word2idx[t]])
    if len(vecs) == 0:
        return np.zeros(embedding_dim)
    total = np.zeros(embedding_dim)
    for v in vecs:
        total = total + v
    return total / len(vecs)
```

```python
# Vector que representa al documento completo
vec_doc = vector_oracion(tokens_global)
print(f"Vector del documento (primeros 5 valores): {vec_doc[:5]}")
```

```python
# Puntuar cada oración contra el vector del documento
puntuaciones = []   # lista de tuplas (indice_oracion, score)

for i in range(len(oraciones_originales)):
    toks  = tokenizador(to_minusculas(oraciones_originales[i]), stopwords)
    vec   = vector_oracion(toks)
    score = similitud_coseno(vec, vec_doc)
    puntuaciones.append((i, score))

# Ordenar de mayor a menor (selection sort descendente)
for i in range(len(puntuaciones) - 1):
    for j in range(i + 1, len(puntuaciones)):
        if puntuaciones[j][1] > puntuaciones[i][1]:
            puntuaciones[i], puntuaciones[j] = puntuaciones[j], puntuaciones[i]

print(f"{'Pos':<5} {'Score':<10} Inicio de la oración")

for rank, (idx, score) in enumerate(puntuaciones):
    print(f"  {rank+1:<4} {score:.4f}     {oraciones_originales[idx][:55]}...")
```

# Resumen final

Se toman las N oraciones con mayor score y se reordenan por su posición original para que el resumen sea narrativamente fluido.

```python
N_RESUMEN = 8  # cantidad de oraciones en el resumen

# Índices de las top-N oraciones
top_indices = [p[0] for p in puntuaciones[:N_RESUMEN]]

# Reordenar las oraciones
# Posición original en el texto
top_indices_ordenados = sorted(top_indices)

print(f"Oraciones seleccionadas (índices): {top_indices_ordenados}")
```

```python
print("Resumen generado con Word2vec:")
print()
resumen = " ".join([oraciones_originales[i] for i in top_indices_ordenados])
print(resumen)

tokens_resumen = tokenizador(to_minusculas(resumen), stopwords)
print(f"\nTokens en el texto original : {len(tokens_global)}")
print(f"Tokens en el resumen        : {len(tokens_resumen)}")
print(f"Tasa de compresión          : {100*(1 - len(tokens_resumen)/len(tokens_global)):.2f}%")
```
