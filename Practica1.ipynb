## A)
Permitir acentos

```python
def es_valido(caracter):
    c = ord(caracter)
    
    if (c >= 97 and c <= 122) or (c >= 65 and c <= 90) or (c >= 48 and c <= 57) or (c == 130) or (c >=160 and c<=163):
        return True
    else:
        return False

def tokenizador4(texto):
    token = ""
    tokens = [None] * len(texto)
    j = 0
    for i in range(len(texto)):
        if es_valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                tokens[j] = token
                j += 1
                token = ''
    return tokens[0:j]


texto = "Ló5s 10 alúmn1os de 1 PLN2 tienen poca imaginación."
print(tokenizador4(texto))
```

## B)
Permitir caracteres especiales

```python
def es_especial(caracter):
    return not es_valido(caracter) and caracter != ' '


def tokenizador2(texto):
    token = ''
    tokens = [None] * len(texto)
    j = 0

    for i in range(len(texto)):
        c = texto[i]

        if es_valido(c):
            token += c
        elif es_especial(c) and token != '':
            token += c
        else:
            if token != "":
                tokens[j] = token
                j += 1
                token = ''

    if token != '':
        tokens[j] = token
        j += 1

    return tokens[0:j]


texto = "Ló5@@s @ 10 alúmn1os__ de 1 PLN2 tienen poca imaginación @ ."
print(tokenizador2(texto))
```
