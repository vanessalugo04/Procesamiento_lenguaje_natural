# **Conteo de palabras**

```python
def contador_palabras(texto):

    contador = 0
    en_palabra = False

    for caracter in texto:

        if caracter != " " and caracter != "\n" and caracter != "\t":

            if not en_palabra:
                contador += 1
                en_palabra = True

        else:
            en_palabra = False

    return contador
```

```python
cantidad = contador_palabras(texto)
print("Cantidad de palabras en el texto original:", cantidad)
```

# **Conversión del texto a minúsculas**

```python
def to_minusculas(texto):

    letras = ""

    for letra in texto:

        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra

        else:
            letras += letra

    return letras
```

```python
texto_minus = to_minusculas(texto)
print(texto_minus)
```

# **Tokenización del texto**

```python
def tokenizador(texto, set_stopwords):

    token = ""
    tokens = [None]*len(texto)
    j = 0

    texto = texto + " "   # evita perder la última palabra

    for i in range(len(texto)):

        if texto[i].isalpha():

            token += texto[i]

        else:

            if token != "":

                if token not in set_stopwords:
                    tokens[j] = token
                    j += 1

                token = ""

    return tokens[0:j]
```

**Stopwords para texto**

```python
stopwords = set([
    'de','la','que','el','en','y','a','los','del','se','las','por','un','para','con','no','una',
    'su','al','lo','como','más','pero','sus','le','ya','o','este','sí','porque','esta','entre',
    'cuando','muy','sin','sobre','también','me','hasta','hay','donde','quien','desde','todo',
    'nos','durante','todos','uno','les','ni','tú','tus','te','ti','tu','vos','vosotros',
    'vosotras','ustedes','ello','ellos','ellas','mío','mía','míos','mías',
    'tuyo','tuya','tuyos','tuyas','suyo','suya','suyos','suyas','nuestro','nuestra',
    'nuestros','nuestras','al','del'
])
```

# **Aplicación del tokenizador al texto prueba**

## Texto Matemático

```python
import fitz

def convertir(archivo):
    archivo_pdf = fitz.open(archivo)
    texto_total = ""  #acumulador total, algo así como el tokens[]

    for i in range(len(archivo_pdf)):
        pagina = archivo_pdf.load_page(i) #se lee cada pagina del texto
        texto_de_la_pagina = pagina.get_text("text") # se extraen los caracteres, es un acumulador temporal como token=''
        texto_total = texto_total + texto_de_la_pagina + "\n" #agregar cada pagina a texto con salto de linea

    archivo_pdf.close() #se deja de leer el pdf xd

    return texto_total #se devuelve la cadena acumulada
```

```python
matematico = convertir('matematico.pdf')
print(matematico)
```

```python
def es_valido_mate(caracter):
    c = ord(caracter)
    if (c >= 97 and c <= 122) or (c >= 65 and c <= 90) or (c == 130) or (c >= 160 and c <= 165) or (c == 130) or (c >=157 and c<=163):
        return True
    if (c >= 48 and c <= 57):
        return True
    if caracter in "+-*/^=_<>!{}[]()&^":
        return True
    return False


def tokenizador_mate(texto, stop_words_matematicas, valido):
    token = ""
    tokens = [None] * len(texto)
    j = 0
    for i in range(len(texto)):
        if valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                if token not in stop_words_matematicas:
                    tokens[j] = token
                    j += 1
                token = ''
                
    return tokens[0:j]
```

```python
tokens_mate = tokenizador_mate(matematico_minuscula, stop_words_matematicas, es_valido_mate)

print("Tokens del texto matemático sin stopwords:")
print(tokens_mate)

print("Cantidad de tokens:", len(tokens_mate))
```

# **Texto Literario**

**Selección de stopwords para el texto literario**

 En este análisis usamos un conjunto de stopwords pensado especialmente
 para el Quijote, que está escrito en español del siglo XVII.

 Quitamos todas las palabras que aparecen mucho
 pero no dicen nada importante sobre la historia — artículos como "el",
 "la", preposiciones como "de", "en", "por", conjunciones, pronombres, etc.

 Lo interesante de este texto es que además tiene arcaísmos, que son
 palabras antiguas que Cervantes usaba y que hoy ya no existen, como
 "della" (de ella), "deste" (de este), "ansí" (así) o "fasta" (hasta).
 Estas también las eliminamos porque aparecen constantemente pero
 no nos ayudan a entender de qué trata la obra.

```python
stopwords_literarias = set([
    #artículos y determinantes
    'de','la','el','en','y','a','los','del','se','las','por','un','una','para','con',
    'al','lo','unos','unas','mi','tu','su','mis','tus','sus',

    #conjunciones y conectores
    'que','si','ni','o','pero','mas','aunque','pues','como','cuando','donde',
    'porque','sino','cuanto','mientras',

    #preposiciones
    'ante','bajo','con','contra','desde','entre','hacia','hasta','para','por',
    'según','sin','sobre','tras',

    #pronombres
    'yo','me','te','nos','os','le','les','lo','los','se',
    'este','esta','esto','ese','esa','eso','aquel','aquella','aquello',
    'ellos','ellas','él','ella',

    #adverbios sin carga semántica
    'más','muy','tan','así','también','ya','aún','aun','no','sí',
    'aquí','allí','ahí','luego','antes','después','entonces','bien',

    #verbos auxiliares y copulativos
    'es','era','fue','ser','sido','ha','han','he','hemos','son',
    'eran','fueron','sea','será','sería','estar','estaba','estuvo',
    'hay','había','hubo',

    #arcaísmos del Quijote
    'ca','fasta','vuesa','vuestra','vuestro','ansí','mesmo','mesma',
    'della','dello','deste','desta','desto','dellos','dellas',
    'agora','non','quel','cuál',
])
```

*Cargamos el PDF con parte del quijote*

```python
# Cargar el PDF del texto literario
texto_literario = convertir("Miguel de Cervantes El Ingenioso Hidalgo Don Quijote de la Mancha-13-30.pdf")

# Verificar que cargó bien
cantidad=contador_palabras(texto_literario)
print("\nCantidad de palabras",cantidad)
```

```python
print("=== TEXTO ORIGINAL ===")
print(texto_literario[:500])
```

*Aplicamos el tokenizador con las stopwords*

```python
texto_literario = to_minusculas(texto_literario)
tokens_literarios = tokenizador(texto_literario, stopwords_literarias)
print(f"\nCantidad de tokens: {len(tokens_literarios)}")
```
