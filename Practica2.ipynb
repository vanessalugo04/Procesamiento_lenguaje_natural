## **Conteo de palabras**

```python
def contador_palabras(texto):

    contador = 0
    en_palabra = False

    for caracter in texto:

        if caracter != " " and caracter != "\n" and caracter != "\t":

            if not en_palabra:
                contador += 1
                en_palabra = True

        else:
            en_palabra = False

    return contador
```

```python
cantidad = contador_palabras(texto)
print("Cantidad de palabras en el texto original:", cantidad)
```

## **Conversión del texto a minúsculas**

```python
def to_minusculas(texto):

    letras = ""

    for letra in texto:

        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra

        else:
            letras += letra

    return letras
```

```python
texto_minus = to_minusculas(texto)
print(texto_minus)
```

## **Tokenización del texto**

```python
stopwords = set([
    'de','la','que','el','en','y','a','los','del','se','las','por','un','para','con','no','una',
    'su','al','lo','como','más','pero','sus','le','ya','o','este','sí','porque','esta','entre',
    'cuando','muy','sin','sobre','también','me','hasta','hay','donde','quien','desde','todo',
    'nos','durante','todos','uno','les','ni','tú','tus','te','ti','tu','vos','vosotros',
    'vosotras','ustedes','ello','ellos','ellas','mío','mía','míos','mías',
    'tuyo','tuya','tuyos','tuyas','suyo','suya','suyos','suyas','nuestro','nuestra',
    'nuestros','nuestras','al','del'
])
```

```python
def es_valido(caracter):
    c = ord(caracter)
    
    if (c >= 97 and c <= 122) or (c >= 65 and c <= 90) or (c >= 48 and c <= 57):
        return True
    if ((c == 225) or (c == 233) or (c == 327) or (c == 243) or (c == 250) or (c == 241)):#unicode
        return True
    else:
        return False
```

```python
def tokenizador(texto, stopwords, valido):
    token = ""
    tokens = [None] * len(texto)
    j = 0
    
    for i in range(len(texto)):
        if valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                if token not in stopwords:
                    tokens[j] = token
                    j += 1
                token = ''
                
    return tokens[0:j]
```

**Stopwords para texto**

## **Aplicación del tokenizador al texto prueba**

```python
tokens = tokenizador(texto_minus, stopwords, es_valido)

print("Tokens del texto literario sin stopwords:")
print(tokens)

print("Cantidad de tokens:", len(tokens))
```

## Texto Matemático

Se rescató un texto del libro Introducción a Procesos Estocásticos, en donde se encuentran diferentes símbolos matemáticos como: funciones, ecuaciones y símbolos. Se seleccionaron de StopWords concectores, pronombres, etc. Así mismo se usa una función para que tokenice de forma correcta, aceptando acentos, símbolos y las ecuaciones para que no se pierda la estrcutura del texto.

```python
import fitz

def convertir(archivo):
    archivo_pdf = fitz.open(archivo)
    texto_total = ""  #acumulador total, algo así como el tokens[]

    for i in range(len(archivo_pdf)):
        pagina = archivo_pdf.load_page(i) #se lee cada pagina del texto
        texto_de_la_pagina = pagina.get_text("text") # se extraen los caracteres, es un acumulador temporal como token=''
        texto_total = texto_total + texto_de_la_pagina + "\n" #agregar cada pagina a texto con salto de linea

    archivo_pdf.close() #se deja de leer el pdf 

    return texto_total #se devuelve la cadena acumulada
```
