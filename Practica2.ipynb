# **Conteo de palabras**

```python
def contador_palabras(texto):

    contador = 0
    en_palabra = False

    for caracter in texto:

        if caracter != " " and caracter != "\n" and caracter != "\t":

            if not en_palabra:
                contador += 1
                en_palabra = True

        else:
            en_palabra = False

    return contador
```

```python
cantidad = contador_palabras(texto)
print("Cantidad de palabras en el texto original:", cantidad)
```

# **ConversiГіn del texto a minГәsculas**

```python
def to_minusculas(texto):

    letras = ""

    for letra in texto:

        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra

        else:
            letras += letra

    return letras
```

```python
texto_minus = to_minusculas(texto)
print(texto_minus)
```

# **TokenizaciГіn del texto**

```python
def tokenizador(texto, set_stopwords):

    token = ""
    tokens = [None]*len(texto)
    j = 0

    texto = texto + " "   # evita perder la Гәltima palabra

    for i in range(len(texto)):

        if texto[i].isalpha():

            token += texto[i]

        else:

            if token != "":

                if token not in set_stopwords:
                    tokens[j] = token
                    j += 1

                token = ""

    return tokens[0:j]
```

**Stopwords para texto**

```python
stopwords = set([
    'de','la','que','el','en','y','a','los','del','se','las','por','un','para','con','no','una',
    'su','al','lo','como','mГЎs','pero','sus','le','ya','o','este','sГӯ','porque','esta','entre',
    'cuando','muy','sin','sobre','tambiГ©n','me','hasta','hay','donde','quien','desde','todo',
    'nos','durante','todos','uno','les','ni','tГә','tus','te','ti','tu','vos','vosotros',
    'vosotras','ustedes','ello','ellos','ellas','mГӯo','mГӯa','mГӯos','mГӯas',
    'tuyo','tuya','tuyos','tuyas','suyo','suya','suyos','suyas','nuestro','nuestra',
    'nuestros','nuestras','al','del'
])
```

## Texto MatemГЎtico

```python
import fitz

def convertir(archivo):
    archivo_pdf = fitz.open(archivo)
    texto_total = ""  #acumulador total, algo asГӯ como el tokens[]

    for i in range(len(archivo_pdf)):
        pagina = archivo_pdf.load_page(i) #se lee cada pagina del texto
        texto_de_la_pagina = pagina.get_text("text") # se extraen los caracteres, es un acumulador temporal como token=''
        texto_total = texto_total + texto_de_la_pagina + "\n" #agregar cada pagina a texto con salto de linea

    archivo_pdf.close() #se deja de leer el pdf xd

    return texto_total #se devuelve la cadena acumulada
```

NOTA: Estamos agregando codigo de unicode por la lectura del pdf y tipos de simbolos matematicos. Unicode es como el ascii pero moderno y que usa python. El rango de unicode fue propuesto a ojo por el tipo de simbolos.     
Se sigue usando ascii para procesar el txt     
Si se quiere pasar a otro lenguaje como c, entonces solo se usa ascii extendido. Se hizo un tokenizador a parte para introducir la funcion es_valido_mate.

```python
def es_valido_mate(caracter):
    c = ord(caracter)
    if (0x1D400 <= c <= 0x1D7FF): #codigos de unicode
        return True
    if (0x2100 <= c <= 0x214F):
        return True
    if (0x2200 <= c <= 0x22FF):
        return True
    if (0x00C0 <= c <= 0x00FF):
        return True
    if ((c >= 97 and c <= 126) or (c >= 65 and c <= 94) or (c >= 40 and c <= 62) or (c == 33) or (c == 38) or 
        (c == 131) or (c == 136) or (c == 140) or (c==147) or (c==150) or (c == 155) or (c == 240) or (c == 241)
        or (c == 246) or (c >= 251 and c <= 253)):
        return True
    else:
        return False

    return False

def tokenizador_mate(texto, stop_words_matematicas, valido):
    token = ""
    tokens = [None] * len(texto)
    j = 0
    for i in range(len(texto)):
        if valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                if token not in stop_words_matematicas:
                    tokens[j] = token
                    j += 1
                token = ''
                
    return tokens[0:j]
```

```python
tokens_mate = tokenizador_mate(matematico_minuscula, stop_words_matematicas, es_valido_mate)

print("Tokens del texto matemГЎtico sin stopwords:")
print(tokens_mate)

print("Cantidad de tokens:", len(tokens_mate))
```

# **Texto Literario**

**SelecciГіn de stopwords para el texto literario**

 En este anГЎlisis usamos un conjunto de stopwords pensado especialmente
 para el Quijote, que estГЎ escrito en espaГұol del siglo XVII.

 Quitamos todas las palabras que aparecen mucho
 pero no dicen nada importante sobre la historia вҖ” artГӯculos como "el",
 "la", preposiciones como "de", "en", "por", conjunciones, pronombres, etc.

 Lo interesante de este texto es que ademГЎs tiene arcaГӯsmos, que son
 palabras antiguas que Cervantes usaba y que hoy ya no existen, como
 "della" (de ella), "deste" (de este), "ansГӯ" (asГӯ) o "fasta" (hasta).
 Estas tambiГ©n las eliminamos porque aparecen constantemente pero
 no nos ayudan a entender de quГ© trata la obra.

```python
stopwords_literarias = set([
    #artГӯculos y determinantes
    'de','la','el','en','y','a','los','del','se','las','por','un','una','para','con',
    'al','lo','unos','unas','mi','tu','su','mis','tus','sus',

    #conjunciones y conectores
    'que','si','ni','o','pero','mas','aunque','pues','como','cuando','donde',
    'porque','sino','cuanto','mientras',

    #preposiciones
    'ante','bajo','con','contra','desde','entre','hacia','hasta','para','por',
    'segГәn','sin','sobre','tras',

    #pronombres
    'yo','me','te','nos','os','le','les','lo','los','se',
    'este','esta','esto','ese','esa','eso','aquel','aquella','aquello',
    'ellos','ellas','Г©l','ella',

    #adverbios sin carga semГЎntica
    'mГЎs','muy','tan','asГӯ','tambiГ©n','ya','aГәn','aun','no','sГӯ',
    'aquГӯ','allГӯ','ahГӯ','luego','antes','despuГ©s','entonces','bien',

    #verbos auxiliares y copulativos
    'es','era','fue','ser','sido','ha','han','he','hemos','son',
    'eran','fueron','sea','serГЎ','serГӯa','estar','estaba','estuvo',
    'hay','habГӯa','hubo',

    #arcaГӯsmos del Quijote
    'ca','fasta','vuesa','vuestra','vuestro','ansГӯ','mesmo','mesma',
    'della','dello','deste','desta','desto','dellos','dellas',
    'agora','non','quel','cuГЎl',
])
```
