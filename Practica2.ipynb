```python
texto = """Frente a Ã©l estaban los miles de libros reciÃ©n editados, y todos con su nombre, no podÃ­a creerlo, al fin era un escritor. SaliÃ³ de la casa editorial y fue con su primer libro donde los seres que mÃ¡s amaba. LlegÃ³ a la casa de su padre y le entregÃ³ el libro. Su padre lo cogiÃ³ y, antes de abrirlo, le preguntÃ³: Â¿QuÃ©? Â¿Eres escritor?... El muchacho no supo quÃ© responderle, pero allÃ­ estaba su nombre, su foto en la dobladillo de la carÃ¡tula, estaba todo, pero sÃ³lo respondiÃ³ que no, que no era suyo sino de un amigo, y que Ã©l habÃ­a usado su nombre, pues no deseaba hacerse conocido, odiaba la notoriedad. Su padre se alegrÃ³ por la noticia, y abriÃ³ el libro y no lo soltÃ³ hasta terminarlo; eso luego de todo el dÃ­a, mientras su hijo se deleitaba viÃ©ndole leerle, hasta quedarse dormido... Un golpe en el hombro lo despertÃ³. Era su madre que con el libro en la mano se lo entregaba... Â¿Y? Â¿QuÃ© dijo papÃ¡?, preguntÃ³. Bueno, tÃº lo conoces, lo dejÃ³ allÃ­, y dejÃ³ unos apuntes; creo que estÃ¡n dentro de las hojas del libro. El muchacho lo abriÃ³ y lo que leyÃ³ no le gustÃ³, pero... era la letra de su padre, y si era asÃ­, eran honestas... "Un libro no maduro", leyÃ³. Se despidiÃ³ de su madre y fue hacia la casa editorial. Estaba cerrada, pero, Ã©l la conocÃ­a muy bien, pues desde que era un chico habÃ­a trabajado en ella. EntrÃ³ por una de las ventanas y sacÃ³ todos sus libros. Los puso en una caja y se los llevÃ³. CaminÃ³ hasta llegar a un puente y, sin pestaÃ±ear, los tirÃ³ al barranco... Mientras caÃ­an como palomas sin alas, el muchacho sintiÃ³ como una liberaciÃ³n, un sentimiento de contento lo embargÃ³. Se dio media vuelta y fue directo a su casa para escribir otro libro con el tÃ­tulo de: El despenador... TratarÃ­a de un hombre que ayuda a morir a los que agonizan, de dos hermanos, de un padre agonizante y le pondrÃ­a algo de emociÃ³n, como balazos, muertes extraÃ±as, noche con sonidos de hojas secas en las pisadas, etc. Una vez acabada la obra, se la entregÃ³ a su padre antes de pedirle a su jefe que le imprimiera su segundo libro. Su padre lo leyÃ³ y, nuevamente, le preguntÃ³ si Ã©l era escritor, y el muchacho le dijo que no...

Todo ocurriÃ³ igual, y aÃºn asÃ­, nuestro amiguito continuÃ³ escribiendo, mostrÃ¡ndosela a su padre, para luego botar sus manuscritos al fondo del puente, y todo continuarÃ­a asÃ­ sino fuera por una enfermedad de su padre... El muchacho no se despegÃ³ hasta que el viejo estaba soltando sus Ãºltimos alientos, y antes de soltar el Ãºltimo, su padre le dijo que le gustarÃ­a que algÃºn dÃ­a se volviera escritor... El viejo muriÃ³, y, sin embargo, el muchacho nunca mÃ¡s volviÃ³ a escribir, tan solo se dedicÃ³ a trabajar en la editorial hasta el fin de sus dÃ­as... Nunca fue feliz, pero, al menos, no volviÃ³ a soÃ±ar en ser un escritor...
"""
def contador_palabras(texto):
    contador = 0
    en_palabra = False
    for caracter in texto:
        if caracter != " " and caracter != "\n" and caracter != "\t":
            if not en_palabra:
                contador += 1      
                en_palabra = True
        else:
            en_palabra = False   
    return contador

cantidad = contador_palabras(texto)
print(f'La cantidad de palabras en el texto original es: {cantidad}')
```

```python
def to_minusculas(texto):
    letras = ""
    for letra in texto:
        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra
        else:
            letras += letra
    return letras

print(to_minusculas(texto))
```

```python
def remueve_stopwords(texto):
    set_stopwords = set([
        'de', 'la', 'que', 'el', 'en', 'y', 'a', 'los', 'del', 'se', 'las', 'por', 'un', 'para', 'con', 'no', 'una',
        'su', 'al', 'lo', 'como', 'mÃ¡s', 'pero', 'sus', 'le', 'ya', 'o', 'este', 'sÃ­', 'porque', 'esta', 'entre',
        'cuando', 'muy', 'sin', 'sobre', 'tambiÃ©n', 'me', 'hasta', 'hay', 'donde', 'quien', 'desde', 'todo',
        'nos', 'durante', 'todos', 'uno', 'les',  
        'ni', "tÃº", "tus", "te", "ti", "tu", "vos", "vosotros", "vosotras", "ustedes", "su", "sus", "ello", "ellos",
        "ellas", "mÃ­o", "mÃ­a", "mÃ­os", "mÃ­as", "tuyo", "tuya", "tuyos", "tuyas", "suyo", "suya", "suyos",
        "suyas", "nuestro", "nuestra", "nuestros", "nuestras",  "al", "del" ])
    token = ""
    tokens = [None]*len(texto)
    j = 0


    if (ord(texto[-1]) >= 65 and ord(texto[-1]) <= 90) or (ord(texto[-1]) >= 97 and ord(texto[-1]) <= 122):
        texto = texto + ' '
    print(texto)

    for i in range(0, len(texto)):

        ascii_acentos = [225, 233, 237, 243, 250,  193, 201, 205, 211, 218,  241, 209]
        if (ord(texto[i]) >= 65 and ord(texto[i]) <= 90) or (ord(texto[i]) >= 97 and ord(texto[i]) <= 122) or (ord(texto[i]) in ascii_acentos):
            token += texto[i]
        else:
          if token != "":
            es_stopword = False
            for palabrita in set_stopwords:
              if token == palabrita:
                es_stopword = True
                break #si se encuentra, para de buscar 
            if es_stopword == False:
              tokens[j] = token
              j += 1
            token = "" # ete limpia el token para la sig palabra

    return tokens[0:j]

texto_limpio = remueve_stopwords(texto)
print(texto_limpio)
print(len(texto_limpio))
```
