```python
#función para leer pdf, solo mandenla llamar :)
from pypdf import PdfReader

def extraer_texto_pdf(ruta_archivo):
    texto_completo = ""

    try:
        reader = PdfReader(ruta_archivo)

        for pagina in reader.pages:
            texto_extraido = pagina.extract_text()
            if texto_extraido:
                texto_completo += texto_extraido + "\n"

        return texto_completo

    except FileNotFoundError:
        return f"Error: No se encontró el archivo '{ruta_archivo}'"
    except Exception as e:
        return f"Ocurrió un error al leer el PDF: {e}"
```

```python
def contador_palabras(texto):

    contador = 0
    en_palabra = False

    for caracter in texto:

        if caracter != " " and caracter != "\n" and caracter != "\t":

            if not en_palabra:
                contador += 1
                en_palabra = True

        else:
            en_palabra = False

    return contador
```

Conversión del texto a minúsculas

```python
def to_minusculas(texto):

    letras = ""

    for letra in texto:

        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra

        else:
            letras += letra

    return letras
```

Tokenización del texto

```python
def es_valido(caracter):
  c = ord(caracter)
    # minusculas y mayusculas
  if (c >= 97 and c <= 122) or (c >= 65 and c <= 90):
    return True
    #
  if ((c == 225) or (c == 233) or (c == 327) or (c == 243) or (c == 250) or (c == 241)):#unicode
    return True
    # minusculas con acento
  if c == 160 or c == 130 or c == 161 or c == 162 or c == 163:
    return True
    # mayusculas con acento
  if c == 181 or c == 144 or c == 214 or c == 224 or c == 233:
    return True
  else:
    return False
```

```python
def tokenizador(texto, stopwords, valido):
    token = ""
    tokens = [None] * len(texto)
    j = 0

    for i in range(len(texto)):
        if valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                if token not in stopwords:
                    tokens[j] = token
                    j += 1
                token = ''

    return tokens[0:j]
```

## WUP SIMILARITY

```python
def wup_similarity(palabra1, palabra2):
  """
  Calcula la similitud Wu & Palmer entre dos palabras usando WordNet
  Retorna 0 si no se puede calcular la similitud
  """
  # si la palabra es igual, regresamos un 1
  if palabra1 == palabra2:
    return 1.0

# wn.synsets() busca en WordNet todos los significados posibles
# lang='spa' le dice que busque en español
  synsets1 = wn.synsets(palabra1, lang='spa')
  synsets2 = wn.synsets(palabra2, lang='spa')

# si la palabra no esta en wn, retorna 0
  if not synsets1 or not synsets2:
    return 0.0

  max_sim = 0.0

# compara TODOS los significados de palabra1 con TODOS los de palabra2
# y se queda con el mayorsito
  for syn1 in synsets1:     # Para cada significado de palabra1
    for syn2 in synsets2:     # Para cada significado de palabra2
      try:
        sim = syn1.wup_similarity(syn2)   # Calcula similitud
        if sim and sim > max_sim:         # Si es mayor que el máximo anterior
          max_sim = sim                   # Actualiza el máximo
      except:
        continue

  return max_sim if max_sim > 0 else 0.0

```

```python
def calcular_promedio_wup(palabra_consulta, corpus_tokenizado):
    """
    Calcula el promedio de la similitud Wu & Palmer entre la palabra consultada
    y cada palabra del corpus tokenizado
    """
    # caso de que el corpus este vaciooo
    if not corpus_tokenizado:
        return 0.0

    # inicializamos variables
    suma_similitudes = 0.0
    contador = 0

    for palabra_corpus in corpus_tokenizado: #recorremos las palabras en el corpus
        sim = wup_similarity(palabra_consulta, palabra_corpus)
        suma_similitudes += sim
        contador += 1

    # si procesamos al menos una palabra
    if contador > 0:
        #sacamos el promediooo
        promedio = suma_similitudes / contador
        return round(promedio, 5)
    else:
        return 0.0
```

### Función lematizadora

```python
def longitud(palabra):
    cantidad_caracteres = 0
    for caracter in palabra:
        cantidad_caracteres = cantidad_caracteres + 1
    return cantidad_caracteres


def termina_en(palabra, sufijo):
    long_p = longitud(palabra)
    long_s = longitud(sufijo)
    if long_s > long_p:
        return False
    return palabra[long_p - long_s:] == sufijo


def lematizador(texto, stopwords):
    texto_lematizado = '' #acumulador del texto
    j = 0 #contador de palabras
    palabra_actual = '' #acumulador por palabra
    
    texto_procesar = texto + ' '

    for caracter in texto_procesar:
        if caracter != " " and caracter != "\n" and caracter != "\t":
            palabra_actual += caracter #se va armando la palabra
            
        else: #si hay salto de linea se termina de armar la palabra y se procesa
            if longitud(palabra_actual) > 0: #esto evita procesar muchos espacios en blanco xd
                palabra_lematizada = None 
                
                for sufijo, terminacion in REGLAS_VERBOS:
                    if longitud(palabra_actual) - longitud(sufijo) >= 3 and termina_en(palabra_actual, sufijo):
                        palabra_lematizada = palabra_actual[:-longitud(sufijo)] + terminacion
                        break

                if palabra_lematizada is None:
                    if palabra_actual in EXCEPCIONES:
                        palabra_lematizada = EXCEPCIONES[palabra_actual]
                    else:
                        for sufijo, reemplazo in REGLAS_MORFOLOGICAS:
                            min_raiz = 4 if sufijo in ('eres', 'es', 's') else 3
                            if longitud(palabra_actual) - longitud(sufijo) >= min_raiz and termina_en(palabra_actual, sufijo):
                                palabra_lematizada = palabra_actual[:-longitud(sufijo)] + reemplazo
                                break

                        if palabra_lematizada is None: #si la palabra vacia no esta lematizada
                            palabra_lematizada = palabra_actual #la pasa tal cual
 
                if j == 0: #si es la primera palabra en el texto
                    texto_lematizado += palabra_lematizada #se agrega sin espacios
                else:
                    texto_lematizado += ' ' + palabra_lematizada #se agrega espacio
                
                j = j + 1
                
                palabra_actual = ''
    
    return texto_lematizado
```

```python
def lematizador_tokenizado(documento_tokenizado):
    texto_lematizado = ""
    j = 0 
    
    for palabra in documento_tokenizado:
        palabra_lematizada = None
        
        for sufijo, terminacion in REGLAS_VERBOS:
            if longitud(palabra) - longitud(sufijo) >= 3 and termina_en(palabra, sufijo):
                palabra_lematizada = palabra[:-longitud(sufijo)] + terminacion
                break

        if palabra_lematizada is None:
            if palabra in EXCEPCIONES:
                palabra_lematizada = EXCEPCIONES[palabra]
            else:
                for sufijo, reemplazo in REGLAS_MORFOLOGICAS:
                    min_raiz = 4 if sufijo in ('eres', 'es', 's') else 3
                    if longitud(palabra) - longitud(sufijo) >= min_raiz and termina_en(palabra, sufijo):
                        palabra_lematizada = palabra[:-longitud(sufijo)] + reemplazo
                        break

        if palabra_lematizada is None: 
            palabra_lematizada = palabra 
         
        if j == 0: #
            texto_lematizado += palabra_lematizada 
        else:
            texto_lematizado += ' ' + palabra_lematizada 
        
        j += 1

    return texto_lematizado
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet

def generar_pdf_documentos(nombre_pdf, documentos):

    estilos = getSampleStyleSheet()
    elementos = []

    for nombre, texto in documentos.items():

        titulo = Paragraph(f"<b>{nombre}</b>")
        contenido = Paragraph(str(texto).replace("\n","<br/>"), estilos['Normal'])

        bloque = [titulo,Spacer(1,10), contenido, Spacer(1,30)]

        elementos = elementos + bloque

    pdf = SimpleDocTemplate(nombre_pdf)
    pdf.build(elementos)

    print("PDF generado:", nombre_pdf)
```

```python
documentos_finales = {
    'Arte': arte_lematizado,
    'Musica': musica_lematizado,
    'Biologia': biologia_lematizado,
    'Historia': historia_lematizado,
    'Matematicas': matematicas_lematizado
}

generar_pdf_documentos("documentos_finales.pdf", documentos_finales)
```
