```python
#### solo por si al rato es util
"""
textos = {
    "DOC1": doc1_medicina,
    "DOC2": doc2_historia,
    "DOC3":   doc3_musica,
}
"""
def contador_palabras(texto):
    contador = 0
    en_palabra = False
    for caracter in texto:
        if caracter != " " and caracter != "\n" and caracter != "\t":
            if not en_palabra:
                contador += 1
                en_palabra = True
        else:
            en_palabra = False
    return contador
# contamos las palabras de cu
print(" ------- Conteo de palabras iniciales -------")
print (f" Palabras en doc1 - medicina : {contador_palabras(doc1_medicina)}")
print (f" Palabras en doc2 - historia : {contador_palabras(doc2_historia)}")
print (f" Palabras en doc3 - musica   : {contador_palabras(doc3_musica)}")
```

```python
# conversion del texto a minusculas
def to_minusculas(texto):
    letras = ""
    for letra in texto:
        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra
        else:
            letras += letra
    return letras
```

```python
##tokenizador del texto

def es_valido(caracter):
  # ojito, los cambié a UTF8 pq python trabaja internamente con este y no ASCII
    c = ord(caracter)

    # Rango estándar: A-Z (65-90) y a-z (97-122)
    if (c >= 65 and c <= 90) or (c >= 97 and c <= 122):
        return True

    # Caracteres especiales del español en Unicode (UTF-8)
    # á=225, é=233, í=237, ó=243, ú=250, ñ=241
    # Á=193, É=201, Í=205, Ó=211, Ú=218, Ñ=209, ü=252, Ü=220
    vocales_tildes_n = [
        225, 233, 237, 243, 250, 241, # minúsculas
        193, 201, 205, 211, 218, 209, # mayúsculas
        252, 220                      # diéresis
    ]

    if c in vocales_tildes_n:
        return True

    return False

def tokenizador(texto, stopwords, valido):
  token = ""
  tokens = [None] * len(texto)
  j = 0

  for i in range(len(texto)):
    if valido(texto[i]):
      token += texto[i]
    else:
      if token != "":
        if token not in stopwords:
          tokens[j] = token
          j += 1
        token = ''

  return tokens[0:j]
```

```python
# # tokenizamos el texto usando la función tokenizador
corpus_doc1 = tokenizador(to_minusculas(doc1_medicina), stopwords_doc1, es_valido)
corpus_doc2 = tokenizador(to_minusculas(doc2_historia), stopwords_doc2, es_valido)
corpus_doc3 = tokenizador(to_minusculas(doc3_musica), stopwords_doc3, es_valido)
print(" ------- Muestra de los tokens -------")
print(corpus_doc3)
```

## A) Obtener una matriz de TF- IDF para todas las palabras del vocabulario

(habiendo quitado el doble espacio o signos de puntuación)

- **TF(t, d)** = frecuencia del término *t* en documento *d* ÷ total de tokens en *d*
- **IDF(t)** = log(N ÷ df(t)), donde N = número de documentos y df = documentos que contienen *t*
- **TF-IDF(t, d)** = TF(t, d) × IDF(t)

```python
# metemos los 3 docs tokenizados en un solo corpus
corpus = [corpus_doc1, corpus_doc2, corpus_doc3]

#
def TF_IDF(corpus, palabra):
    contador = 0
    num_doc_con_palabra = 0 #Contador para IDF. En cuantos documentos aparece la palabra
    num_doc_corpus= len(corpus) #Cantidad de documentos en corpus
    num_palabras_doc = [None]*num_doc_corpus #Lista de nulos del tamaño de cada documento
    num_coicidencias = [None]*num_doc_corpus #Cuantas veces se repite la palabra en cada documento
    TF = np.zeros(len(corpus), dtype=float)  #Resultado de TF


    ## Calcular el numero de palabras presentes en cada Doc
    for i in range(num_doc_corpus):
      ## Total de palabras presente en cada Doc
      num_palabras_doc[i]=len(corpus[i])
      ## Cuantas veces se repite la palabra en el documento
      for pal in corpus[i]:
        if pal == palabra: #si la palabra actual es igual a la palabra que se busca
          contador = contador + 1 #aumenta el contador
      if contador != 0: #si la palabra aparecio al menos una vez
        num_doc_con_palabra +=1 #suma uno
      num_coicidencias[i] = contador #se guarda las veces que aparece
      contador = 0
      TF[i] = num_coicidencias[i] / num_palabras_doc[i] #Formula de TF
    IDF = np.log(num_doc_corpus / num_doc_con_palabra) #Formula de IDF
    #TF = np.array(TF)
    TF_IDF = TF * IDF #Formula de TF-IDF
    return TF, IDF, TF_IDF

```

```python
# ── Construir vocabulario con ciclos y posiciones ────────────
# Primero contamos cuántas palabras únicas hay como máximo
total_tokens = 0
for i in range(len(corpus)):
    total_tokens += len(corpus[i])

vocabulario_temp = [None] * total_tokens   # arreglo con espacio máximo
tam_vocab = 0                              # cuántas palabras únicas llevamos

for i in range(len(corpus)):                       # recorrer cada documento
    for j in range(len(corpus[i])):                # recorrer cada token
        token = corpus[i][j]
        ya_esta = False
        for k in range(tam_vocab):                 # buscar si ya está en vocabulario
            if vocabulario_temp[k] == token:
                ya_esta = True
                break
        if not ya_esta:
            vocabulario_temp[tam_vocab] = token    # agregar por posición
            tam_vocab += 1

# Recortar al tamaño real y ordenar alfabéticamente
#vocabulario = sorted(vocabulario_temp[:tam_vocab])
# solo recortamos al tamaño real
vocabulario = vocabulario_temp[:tam_vocab]

print(f"Tamaño del vocabulario: {len(vocabulario)} palabras únicas")
print(f"Primeras 15: {vocabulario[:15]}")
```

```python
def mostrar_matriz_filtrada(vocabulario, corpus):
    # Encabezado con anchos fijos (25 para palabra, 30 para cada doc)
    print(f"{'PALABRA':<25} {'DOC1':<30} {'DOC2':<30} {'DOC3':<30}")
    print("-" * 115)

    contador_palabras = 0

    for palabra in vocabulario:
        # Obtenemos los cálculos de tu función TF_IDF
        tf_vals, idf_val, tfidf_vals = TF_IDF(corpus, palabra)

        # Filtro: Solo mostramos si la palabra tiene relevancia en algún documento
        if np.sum(tfidf_vals) > 0:
            # Palabra alineada a la izquierda
            fila = f"{palabra:<25}"

            for i in range(len(corpus)):
                # Formato :.4f para asegurar que veas el 0.0047 y no 0.0
                celda = f"({tf_vals[i]:.4f}, {idf_val:.4f}, {tfidf_vals[i]:.4f})"
                fila += f"{celda:<30}"

            print(fila)
            contador_palabras += 1

    print("-" * 115)
    print(f"Total de palabras con peso mayor a cero: {contador_palabras}")
```

```python
# los valores de las palabras que se quitaron
def mostrar_palabras_removidas(vocabulario, corpus):
    print(f"{'PALABRA REMOVIDA':<25} {'DOC1':<30} {'DOC2':<30} {'DOC3':<30}")
    print("-" * 115)

    contador_removidas = 0

    for palabra in vocabulario:
        tf_vals, idf_val, tfidf_vals = TF_IDF(corpus, palabra)

        # La lógica inversa: Si la suma de TF-IDF es exactamente 0
        if np.sum(tfidf_vals) == 0:
            fila = f"{palabra:<25}"

            for i in range(len(corpus)):
                # Aquí verás por qué se eliminó (generalmente porque el IDF es 0.0000)
                celda = f"({tf_vals[i]:.4f}, {idf_val:.4f}, {tfidf_vals[i]:.4f})"
                fila += f"{celda:<30}"

            print(fila)
            contador_removidas += 1

    print("-" * 115)
    print(f"Total de palabras removidas (sin peso): {contador_removidas}")

mostrar_palabras_removidas(vocabulario, corpus)
```

## C) Obtener las palabras más significativas que representen a los 3 documentos, observar el comportamiento

```python
TOP_N      = 15
num_vocab  = len(vocabulario)
num_docs   = len(corpus)

# ── Calcular toda la matriz TF-IDF en arreglos paralelos ─────
# palabras_idx[p]  → índice de la palabra p en vocabulario
# tfidf_matrix[i][p] → tfidf de palabra p en documento i
tfidf_matrix = [[0.0] * num_vocab for _ in range(num_docs)]

for p in range(num_vocab):
    tf_vals, idf_val, tfidf_vals = TF_IDF(corpus, vocabulario[p])
    for i in range(num_docs):
        tfidf_matrix[i][p] = tfidf_vals[i]

# ── Para cada documento: selection sort descendente ──────────
doc_nombres = ["DOC1 — Medicina", "DOC2 — Historia", "DOC3 — Música"]

for i in range(num_docs):
    print(f"\n{'='*56}")
    print(f"  {doc_nombres[i]}")
    print(f"{'='*56}")
    print(f"  {'Pos':<5} {'Palabra':<22} {'TF':>7}  {'IDF':>7}  {'TF-IDF':>9}")
    print(f"  {'─'*52}")

    # Copiar scores e índices del documento i
    scores  = [0.0]    * num_vocab
    indices = [0]      * num_vocab
    for p in range(num_vocab):
        scores[p]  = tfidf_matrix[i][p]
        indices[p] = p

    # Selection sort descendente por score (ciclos y posiciones)
    for p in range(num_vocab - 1):
        max_pos = p
        for q in range(p + 1, num_vocab):
            if scores[q] > scores[max_pos]:
                max_pos = q
        # intercambiar scores
        scores[p],  scores[max_pos]  = scores[max_pos],  scores[p]
        indices[p], indices[max_pos] = indices[max_pos], indices[p]

    # Mostrar Top-N
    for rank in range(TOP_N):
        idx     = indices[rank]
        palabra = vocabulario[idx]
        tf_vals, idf_val, tfidf_val = TF_IDF(corpus, palabra)
        print(f"  {rank+1:<5} {palabra:<22} {round(tf_vals[i],4):>7.4f}  "
              f"{round(idf_val,4):>7.4f}  {round(tfidf_val[i],6):>9.6f}")
```
