### Cargar PDF de doc original, plagio y no plagio 

```python
from ipywidgets import FileUpload
from IPython.display import display

print("Sube el PDF que crees que sea el original :")
upload_original = FileUpload(accept='.pdf', multiple=False)
display(upload_original)

print("\nsube el PDF que crees que tenga plagio:")
upload_plagio = FileUpload(accept='.pdf', multiple=False)
display(upload_plagio)

print("\nsube el PDF que crees que no tenga plagio:")
upload_no_plagio = FileUpload(accept='.pdf', multiple=False)
display(upload_no_plagio)
```

```python
from PyPDF2 import PdfReader
import io

def leer_pdf_bytes(file_bytes):
    texto = ""
    reader = PdfReader(io.BytesIO(file_bytes))
    
    for pagina in reader.pages:
        contenido = pagina.extract_text()
        if contenido:
            texto += contenido + "\n"
    
    return texto
    
archivo_original = upload_original.value[0]
archivo_plagio = upload_plagio.value[0]
archivo_no_plagio = upload_plagio.value[0]

# extraer contenido
bytes_original = archivo_original['content']
bytes_plagio = archivo_plagio['content']
bytes_no_plagio = archivo_plagio['content']

# convertir a texto
texto_original = leer_pdf_bytes(bytes_original)
texto_plagiado = leer_pdf_bytes(bytes_plagio)
texto_no_plagiado = leer_pdf_bytes(bytes_no_plagio)

print("Archivos cargados correctamente")
```

### Contador de palabras

```python
def contador_palabras(texto):

    contador = 0
    en_palabra = False

    for caracter in texto:

        if caracter != " " and caracter != "\n" and caracter != "\t":

            if not en_palabra:
                contador += 1
                en_palabra = True

        else:
            en_palabra = False

    return contador
```

```python
cantidad_texto_original = contador_palabras(texto_original)
cantidad_texto_plagiado = contador_palabras(texto_plagiado)
cantidad_texto_no_plagiado = contador_palabras(texto_no_plagiado)

print("Cantidad de palabras en el texto original:", cantidad_texto_original)
print("Cantidad de palabras en el texto plagiado:", cantidad_texto_plagiado)
print("Cantidad de palabras en el texto no plagiado:", cantidad_texto_no_plagiado)
```

### Conversión a minusculas

```python
def to_minusculas(texto):

    letras = ""

    for letra in texto:

        if ord(letra) >= 65 and ord(letra) <= 90:
            nueva_letra = chr(ord(letra) + 32)
            letras += nueva_letra

        else:
            letras += letra

    return letras
```

```python
minus_texto_original = to_minusculas(texto_original)
minus_texto_plagiado = to_minusculas(texto_plagiado)
minus_texto_no_plagiado = to_minusculas(texto_no_plagiado)
```

### Tokenización

```python
stopwords = ([
    'a','all','also','although','among','an','and','are','around','as','at', 'be','been','being','both','but','by',
    'can','continue','despite','do','due', 'each', 'for','from','general', 'has','have','however',
    'in','include','including','into','is','it','its','like', 'many','more','most','must', 'nevertheless',
    'of','on','one','or','other','overall', 'plays','provide','provides', 'remains', 'such',
    'that','the','their','these','this','to','under','various','well','with'
])
```

```python
def es_valido(caracter):
  c = ord(caracter)
    # minusculas y mayusculas
  if (c >= 97 and c <= 122) or (c >= 65 and c <= 90):
    return True
    #
  if ((c == 225) or (c == 233) or (c == 327) or (c == 243) or (c == 250) or (c == 241)):#unicode
    return True
    # minusculas con acento
  if c == 160 or c == 130 or c == 161 or c == 162 or c == 163:
    return True
    # mayusculas con acento
  if c == 181 or c == 144 or c == 214 or c == 224 or c == 233:
    return True
  else:
    return False
```

```python
def tokenizador(texto, stopwords, valido):
    token = ""
    tokens = [None] * len(texto)
    j = 0

    for i in range(len(texto)):
        if valido(texto[i]):
            token += texto[i]
        else:
            if token != "":
                if token not in stopwords:
                    tokens[j] = token
                    j += 1
                token = ''

    return tokens[0:j]
```

```python
def contar_tokens(lista_tokens):
    contador = 0
    for elemento in lista_tokens:
        contador += 1
    return contador
```

```python
texto_original_tokenizado = tokenizador(minus_texto_original, stopwords, es_valido)
texto_plagiado_tokenizado = tokenizador(minus_texto_plagiado, stopwords, es_valido)
texto_no_plagiado_tokenizado = tokenizador(minus_texto_no_plagiado, stopwords, es_valido)

print(f"\n-------------------Tokens Texto original \n {texto_original_tokenizado}")
print(f"\n-------------------Tokens Texto Plagiado \n {texto_plagiado_tokenizado}")
print(f"\n------------------- Tokens Texto no Plagiado \n {texto_no_plagiado_tokenizado}")
```

```python
print(f"\nCantidad de tokens texto original : {contar_tokens(texto_original_tokenizado)}")
print(f"\nCantidad de tokens texto plagiado : {contar_tokens(texto_plagiado_tokenizado)}")
print(f"\nCantidad de tokens texto no plagiado : {contar_tokens(texto_no_plagiado_tokenizado)}")
```
